```python
# MODEL TO DETECT
#Using PlainTextCorpusReader
from nltk.corpus import PlaintextCorpusReader
from nltk import bigrams,trigrams
from collections import Counter, defaultdict

corpus_root= r'your *data* folder directory path. Check ReadMe If you need details. Example: C:\Users\PA\Documents\IGBO Stuffs\data'
wordlists= PlaintextCorpusReader(corpus_root,'.*')
#wordlists.fileids()
#wordlists.words()
#wordlists.sents()

def _my_normalization():
    return defaultdict(int)

#placeholder for model
model_Tri_Center= defaultdict(_my_normalization)
model_Tri_Left= defaultdict(_my_normalization)
model_Tri_Right= defaultdict(_my_normalization)
model_Bi_Left= defaultdict(_my_normalization)
model_Bi_Right= defaultdict(_my_normalization)

""" I can adjust the direction of my n-grams by adjusting positions in (a and b) below accordingly for now, I stick with
    the models I desire to use. Which are Trigram right, Bigram Right, Trigram center, Bigram left, Trigram left
    
    1. Trigram to the right will cover for words within the test or training data, which occurs in the middle of the text
       i.e not 1st or 2nd word. 
    2. Trigram left or bigram left could help predict the 1st word error as we can obtain both words or the next word after
       the 1st word to predict the probability of that word occuring
    3. Trigram center or bigram right can solve for the 2nd word if it is an erro"""

#count frequency of co-occurrence....a
"""
# TRIGRAMS"""
for sentence in wordlists.sents():
    for w1,w2,w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model_Tri_Right[(w1,w2)][w3] +=1 #Trigram Right model
        model_Tri_Center[(w1,w3)][w2] +=1 #Trigram center model
        model_Tri_Left[(w2,w3)][w1] +=1 #Trigram left model
        

""" #BIGRAMS"""
for sentence in wordlists.sents():
    for w1,w2 in bigrams(sentence, pad_right=True, pad_left=True):
        model_Bi_Right[w1][w2] +=1 #bigram right model
        model_Bi_Left[w2][w1] +=1 #bigram left model
        
#Transform into probabilities...b

#TRIGRAMS

"""
#Trigram Right Model"""
for w1_w2 in model_Tri_Right:
    total_count= float(sum(model_Tri_Right[w1_w2].values()))
    for w3 in model_Tri_Right[w1_w2]:
        model_Tri_Right[w1_w2][w3] /= total_count

"""
#Trigram center Model"""
for w1_w3 in model_Tri_Center:
    total_count= float(sum(model_Tri_Center[w1_w3].values()))
    for w2 in model_Tri_Center[w1_w3]:
        model_Tri_Center[w1_w3][w2] /= total_count

"""
#Trigram left Model"""
for w2_w3 in model_Tri_Left:
    total_count= float(sum(model_Tri_Left[w2_w3].values()))
    for w1 in model_Tri_Left[w2_w3]:
        model_Tri_Left[w2_w3][w1] /= total_count

#BIGRAMS
        
"""Bigram Right Model"""
for w1 in model_Bi_Right:
    total_count= float(sum(model_Bi_Right[w1].values()))
    for w2 in model_Bi_Right[w1]:
        model_Bi_Right[w1][w2] /= total_count
        
"""Bigram Right Model"""
for w2 in model_Bi_Left:
    total_count= float(sum(model_Bi_Left[w2].values()))
    for w1 in model_Bi_Left[w2]:
        model_Bi_Left[w2][w1] /= total_count


# TRY RUNNING ALL OF THEM IN A NEW TERMINAL, AND USING SEPERATE PLACE HOLDERS FOR EACH OF THEM
```

```python
import pickle 
dictionary_Tri_Left = model_Tri_Left
dictionary_Tri_Right = model_Tri_Right
dictionary_Tri_Center = model_Tri_Center
dictionary_Bi_Left = model_Bi_Left
dictionary_Bi_Right = model_Bi_Right



#print(dict(model_Tri_Left["site","naka"]))

# SAVE
with open("Tri_Left.pkl", "wb") as pkl_handle:
	pickle.dump(dictionary_Tri_Left, pkl_handle)
    
with open("Tri_Right.pkl", "wb") as pkl_handle:
	pickle.dump(dictionary_Tri_Right, pkl_handle)
    
with open("Tri_Center.pkl", "wb") as pkl_handle:
	pickle.dump(dictionary_Tri_Center, pkl_handle)
    
with open("Bi_Left.pkl", "wb") as pkl_handle:
	pickle.dump(dictionary_Bi_Left, pkl_handle)
    
with open("Bi_Right.pkl", "wb") as pkl_handle:
	pickle.dump(dictionary_Bi_Right, pkl_handle)
    
# LOAD
with open("Tri_Left.pkl", "rb") as pkl_handle:
	model_Tri_Left_A = pickle.load(pkl_handle)
    
#print(dict(model_Tri_Left_A["site","naka"]))

with open("Tri_Right.pkl", "rb") as pkl_handle:
	model_Tri_Right_A = pickle.load(pkl_handle)

#print(dict(model_Tri_Right_A["site","naka"]))

with open("Bi_Right.pkl", "rb") as pkl_handle:
	model_Bi_Right_A = pickle.load(pkl_handle)
    
#print (dict(model_Bi_Right_A["site","naka"]))
```
