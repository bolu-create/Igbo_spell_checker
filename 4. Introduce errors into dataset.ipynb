## Code to introduce errors into dataset

```python
import random

# Function to introduce errors to a word
def add_errors(word):
    # Choose a random error type
    error_type = random.choice(["insertion", "deletion", "substitution", "transposition"])

    # Insertion error
    if error_type == "insertion":
        index = random.randint(0, len(word))
        char = random.choice(list("abcdefghijklmnopqrstuvwxyz"))
        return word[:index] + char + word[index:]

    # Deletion error
    elif error_type == "deletion":
        index = random.randint(0, len(word) - 1)
        return word[:index] + word[index+1:]

    # Substitution error
    elif error_type == "substitution":
        index = random.randint(0, len(word) - 1)
        char = random.choice(list("abcdefghijklmnopqrstuvwxyz"))
        return word[:index] + char + word[index+1:]

    # Transposition error
    elif error_type == "transposition":
        index = random.randint(0, len(word) - 2)
        return word[:index] + word[index+1] + word[index] + word[index+2:]

# Open the text file and read the contents
with open("test.ig.txt", "r",encoding="utf8") as f:
    lines = f.readlines()

# Split the lines into words
words_list = []
for line in lines:
    words = line.split()
    words_list.append(words)

# Select 30% of the words at random
num_words = sum(len(words) for words in words_list)
num_selected = int(num_words * 0.3)
selected_indices = random.sample(range(num_words), num_selected)

# Loop through the selected words and add errors
count = 0
for i, words in enumerate(words_list):
    for j, word in enumerate(words):
        if count in selected_indices and len(word) > 1:
            new_word = add_errors(word)
            if new_word[0] == word[0]:
                words_list[i][j] = new_word
        count += 1

# Join the modified words back into lines
new_lines = []
for words in words_list:
    new_line = " ".join(words)
    new_lines.append(new_line)

# Join the modified lines back into a single string
new_text = "\n".join(new_lines)

# Write the modified text back to the file
with open("text_file_modified.txt", "w",encoding="utf8") as f:
    f.write(new_text)
```

```python
import random
import string

# Pre-generate a list of random characters for substitution and insertion errors
char_list = list(string.ascii_lowercase)

# Function to introduce errors to a word
def add_errors(word, char_list):
    # Choose a random error type
    error_type = random.choice(["insertion", "deletion", "substitution", "transposition"])

    # Insertion error
    if error_type == "insertion":
        index = random.randint(0, len(word))
        char = random.choice(char_list)
        return word[:index] + char + word[index:]

    # Deletion error
    elif error_type == "deletion":
        index = random.randint(0, len(word) - 1)
        return word[:index] + word[index+1:]

    # Substitution error
    elif error_type == "substitution":
        index = random.randint(0, len(word) - 1)
        char = random.choice(char_list)
        return word[:index] + char + word[index+1:]

    # Transposition error
    elif error_type == "transposition":
        index = random.randint(0, len(word) - 2)
        return word[:index] + word[index+1] + word[index] + word[index+2:]

# Open the text file and read the contents line by line
with open("test.ig.txt", "r", encoding="utf8") as f, open("text_file_modified.txt", "w",encoding="utf8") as f_modified:
    for line in f:
        # Split the line into words
        words = line.split()

        # Select 30% of the words at random
        num_selected = int(len(words) * 0.3)
        selected_indices = random.sample(range(len(words)), num_selected)

        # Loop through the selected words and add errors
        for j, word in enumerate(words):
            if j in selected_indices and len(word) > 1:
                new_word = add_errors(word, char_list)
                if new_word[0] == word[0]:
                    words[j] = new_word

        # Write the modified line back to the file
        f_modified.write(" ".join(words) + '\n')
```
